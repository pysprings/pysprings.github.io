```python
%matplotlib inline
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sklearn
from scipy import signal
import imageio
from IPython.display import Image
```

```python
sobel_right = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
sobel_bottom_right = np.array([[1, 1, 0],[1, 0, -1], [0, -1, -1]])
sobel_top = sobel_right.T
```

```python
smiley = imageio.imread('smiley_face.png')
squares = imageio.imread('squares.png')
Image(filename='smiley_face.png')
plt.imshow(squares[:, :, 3])
squares = squares[:, :, 3]
plt.imshow(squares)
squares.shape
```

```python
right_grad = signal.convolve2d(smiley, sobel_right)
angle_grad = signal.convolve2d(smiley, sobel_bottom_right)
```

```python
plt.imshow(right_grad)
plt.show()
plt.imshow(angle_grad)
```

```python
right_grad_sq = signal.convolve2d(squares, sobel_right)
top_grad_sq = signal.convolve2d(squares, sobel_top)
angle_grad_sq = signal.convolve2d(squares, sobel_bottom_right)
plt.imshow(right_grad_sq)
plt.show()
plt.imshow(angle_grad_sq)
plt.show()
plt.imshow(top_grad_sq)
```
